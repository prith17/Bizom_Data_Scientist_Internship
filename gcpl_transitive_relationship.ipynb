```python
# Alternatively, you can use the to_numpy() method
src_array = gcpl_filtered_df['src'].to_numpy()
ref_array = gcpl_filtered_df['ref'].to_numpy()
print(src_array,ref_array)
```

```python
def create_symmetric_relations(x,y,dict):
    # for i in range(len(src_array)):
        if x not in dict:
            dict[x] = []
        if y not in dict:
            dict[y] = []
        if y not in dict[x]:
            dict[x].append(y)
        if x not in dict[y]:
            dict[y].append(x)    
        return dict
```

```python
dict={}
for i in range(len(src_array)): 
    x=src_array[i] 
    y=ref_array[i]
    dict=create_symmetric_relations(x,y,dict)
```

```python
for i,k_array in dict.items():
    for k in k_array:
        if k in dict:
            j_array=dict[k]
            for j in j_array:
                create_symmetric_relations(i,j,dict)
```
